```python
def TestExecTime(teststring,catCheck=False):
  command = ["./vm","puzzle1"] if not catCheck else ["cat","-"]
  timeStarted = time.time()
  result = subprocess.run(command,input=teststring, encoding="ascii",capture_output=True)
  endtime = time.time() - timeStarted
  return result,endtime
```

```python
from ast import keyword
baremin = "hack{}"
template = "hack{REPLACE}"
knownsizekey = "hack{aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa}"
lenToGen = len("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa")
def getStringFromChunks(width3indices,width2index):
  assert len(width3indices) == 13
  assert width2index<len(chunksize2)
  assert all([i<len(chunksize3) for i in width3indices])
  w3i = width3indices
  return (chunksize3[w3i[0]]
    + chunksize3[w3i[1]]
    + chunksize3[w3i[2]]
    + chunksize3[w3i[3]]
    + chunksize3[w3i[4]]
    + chunksize3[w3i[5]]
    + chunksize3[w3i[6]]
    + chunksize3[w3i[7]]
    + chunksize3[w3i[8]]
    + chunksize3[w3i[9]]
    + chunksize3[w3i[10]]
    + chunksize3[w3i[11]]
    + chunksize3[w3i[12]]
    + chunksize2[width2index]
    )


width3startingindices = [0]*13
width2chunkindex = 0
key = template.replace("REPLACE",getStringFromChunks(width3startingindices,width2chunkindex))
assert(len(key) == len(knownsizekey))

width3chunkindices = width3startingindices.copy()

keydict = {}

for width2chunkindex in range(0,len(chunksize2)):
  for i in range(0,13):
    for j in range(0,len(chunksize3)):
      width3chunkindices[i] = j
      key = template.replace("REPLACE",getStringFromChunks(width3chunkindices, width2chunkindex))
      assert len(key) == len(knownsizekey)
      # print(key)
      r1,t1 = TestExecTime(key)
      r2,t2 = TestExecTime(key)
      keydict[key] = (t1+t2)/2
      print(f"{(j/len(chunksize3))*100}% done")
    break
  break
```
